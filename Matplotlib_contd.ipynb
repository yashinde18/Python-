```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 5, 8, 6])

plt.bar(x, y)           #if you wat the bars to be displayed horizontally instead of vertically, use the barh(x, y)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 5, 8, 6])

plt.bar(x, y, color = "red")
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 5, 8, 6])

plt.bar(x, y, width = 0.2)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 5, 8, 6])

plt.barh(x, y, height = 0.2)
plt.show()
```

```python
x = np.random.normal(170, 10, 250)    # 170 is mean, 10 is std deviation, 250 is no. of observation

plt.hist(x)
plt.show()
```

```python
y = np.array([46, 6,65,78])
plt.pie(y)
plt.show()
```

```python
y = np.array([46, 6,65,78])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

```python
y = np.array([46, 6,65,78])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

```python
y = np.array([46, 6,65,78])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0.8, 0.1, 0]    # distance from centre

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

```python
y = np.array([46, 6,65,78])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "blue", "green"]


plt.pie(y, labels = mylabels, colors = mycolors)
plt.show()
```

```python
y = np.array([46, 6,65,78])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show()
```
