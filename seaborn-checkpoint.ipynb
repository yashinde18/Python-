```python
# import the dependencies and seaborn
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import scipy as stats
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Premalatha-success/Statistics/main/CardioGoodFitness-1.csv")
```

```python
df.head()
```

```python
sns.relplot(x = "Age", y = "Income", data = df)
```

```python
sns.relplot(x = "Age", y = "Income", hue = "Gender", data = df)     #different colors because hue here is categorical data
```

```python
sns.relplot(x = "Age", y = "Income", hue = "Education", data = df)    # sam color because hue here is numerical data
```

```python

sns.relplot(x="Age", y="Income", size = "Education", data = df)
```

```python
sns.catplot(x = "Gender", y = "Income", data = df)
```

```python
sns.catplot(x = "MaritalStatus", y = "Miles", jitter = False, data = df)      # jitter give no scattering
```

```python

sns.catplot(x = "MaritalStatus", y = "Age", hue = "Gender", data = df)
```

```python
sns.catplot(x = "Gender", y = "Income", hue = "MaritalStatus", kind = "swarm", data = df)   

#swarm prevents the points from overlapping 
```

```python
sns.catplot(x = "Gender", y = "Income", hue = "MaritalStatus", kind = "box", data = df)    # boxplot

# dotted line is outlier values
```

```python
sns.catplot(x = "Gender", y = "Income", hue = "MaritalStatus", kind = "violin", data = df)     # violin
```

```python
sns.catplot(x = "Gender", y = "Income", hue = "MaritalStatus", kind = "violin", split = True, data = df)
```

```python
 sns.catplot(x = "Gender", y = "Income", hue = "MaritalStatus", kind = "bar", data = df)    # boxplot using seaborn
```

```python
sns.catplot(x = "MaritalStatus", y = "Income",hue = "Gender", kind = "point", data = df)

# pointplot using seaborn: Pointplot connects data from the same hue category. 
# This helps in identifying how the relationship is changing in a particular hue category.
```

```python

sns.catplot(x = "Gender", y = "Income", hue = "MaritalStatus",
            col = "Product", aspect = 1,
            kind =  "swarm", data = df); 

# col : column
```

```python
sns.displot(df.Age, kde = True)

# distplot() function draws histogram and fits a Kernel Density Estimate.
```

```python
sns.distplot(df.Income, kde = False, rug = True)

# rugplot:  at every observation, it will draw a small vertical stick.
```

```python
sns.jointplot(x= "Age", y = "Income", data = df)

# jointplot by default draws a scatter plot.
# we can see how two independent variables are distributed with respect to each other.
```

```python
sns.jointplot(x = "Age", y = "Income", kind = "hex", data = df)

# hexplot: it shows the number of observations that falls within hexagonal bins. 
# This is a plot which works with large dataset very easily. 
```

```python
sns.jointplot(x = "Age", y = "Income", kind = "kde", data = df)
```

```python
corrmat = df.corr()
f, ax = plt.subplots(figsize = (9, 6))
sns.heatmap(corrmat, vmax = 1, square = True)

# heatmaps using seaborn
```

```python
 sns.catplot(x = "Age", y = "Income", hue = "Gender", kind = "boxen", height = 4, aspect = 2.7, data = df) 
    
# boxen shows large number of values of a variable, also known as quantiles
```

```python
sns.pairplot(df)

# This shows the relationship between each column of the database. 
# It also draws the univariate distribution plot of each variable on the diagonal axis. 
```
