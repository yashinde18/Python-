```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
xpoints = np.array([0, 10])
ypoints = np.array([5, 150])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([1, 2])
ypoints = np.array([10, 20])

plt.plot(xpoints, ypoints, '+')
plt.show()
```

```python
x = np.array([40, 400])
y = np.array([60, 600])

plt.plot(x, y, marker = 'o')
plt.show
```

```python
y = np.array([5,6,76,89])
plt.plot(y, 'o--g')
```

```python
plt.plot(y, marker = 'o', ms = 30, mec = 'b', mfc = 'r')     ### ms is marker size,  mec is marker edge color,  mfc = marker color
plt.show()
```

```python
plt.plot(y, linestyle = 'dashed',marker = 'o')
plt.show()
```

```python
plt.plot(y, ls = ':')
plt.show()
```

```python
plt.plot(y, linewidth = '10')
plt.show()
```

```python
y1 = np.array([3,6,8,2])
y2 = np.array([1,6,4,10])
plt.plot(y1)
plt.plot(y2)
plt.show()
```

```python
x1 = np.array([3,7,2,11])
y1 = np.array([3,6,8,2])
x2 = np.array([7,3,5,8])
y2 = np.array([1,6,4,10])

plt.plot(x1,y1,x2,y2)

plt.grid()     #### plt.grid(axis = 'x')

plt.title('Sports Watch Data' , loc = 'center')
plt.xlabel('Average Pulse')
plt.ylabel('Calorie Burnage')
plt.show()
```

```python
x = np.array([0,1,2,3])
y = np.array([3,5,7,9])

plt.subplot(1,2,1)               ### arguments : 1=rows, 2=columns, 3=dataset number
plt.plot(x,y)


x = np.array([0,1,2,3])
y = np.array([3,34,25,12])

plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
```
